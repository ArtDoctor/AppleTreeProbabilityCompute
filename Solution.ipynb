# Online Math Olympiad Apple Tree Probability Computational Solution
This notebook is a computational approximate solution to the "Apple Tree Probability" problem stated in MindYourDecisions video, https://www.youtube.com/watch?v=RY7YKSw1t_M.

```python
# Needed libraries: random seeds growing, plots and stylized plots
import random
import numpy as np
import matplotlib.pyplot as plt
# from jupyterthemes import jtplot
# jtplot.style(theme='onedork', context='notebook', ticks=True, grid=False)
plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi'] = 100
```

```python
# 1 experiment with 6 seeds
seeds = 6
days = 0
# While not all seeds have grown, continue
while seeds > 0:
    # Iteration through every day
    days += 1
    for seed in range(0, seeds):
        # If random value in range [0,1] is bigger than 0.5, seed grows
        if random.random() > 0.5:
            # Seed is a tree now, amount of seeds left is 1 less
            seeds -= 1
print(days)
```

```python
# Calculate amount average amount of days, using [length] experiments
days = 0
# Do 100000 experiments
length = 100000
for k in range(0, length):
    seeds = 6
    while seeds > 0:
        days+=1
        for seed in range(0, seeds):
            if random.random() > 0.5:
                seeds -= 1
# Calculate the average amount of days
days = days/length
print(days)
```

```python
# Calculate average amount of days for different amount of seeds
res = []
seedVal = 1
accuracy = 100
for j in range(0, 20):
    print("Seeds[",j,"]: ", seedVal)
    days = 0
    for k in range(0, accuracy):
        seeds = seedVal
        while seeds > 0:
            days += 1
            for seed in range(0, seeds):
                if random.random() > 0.5:
                    seeds -= 1
    days = days / accuracy
    res += [days]
    # Increase seed amount
    seedVal *= 2
print(res)
```

```python
# Plot the dependancy
seedVals = 2**np.arange(len(res))
plt.plot(seedVals, res)
plt.plot(seedVals, np.log2(seedVals)+1.4)
plt.xlabel('SeedAmount, N')
plt.ylabel('AmountOfDays, E')
plt.title('E(N)')
plt.show()
```

We can see that the graph looks very same with function E(N) = log2(N) + a, where a is approximately 1.4 (could be sqrt(2) possibly).
